# Replace Salary values ['<=50k','>50k'] with 0 and 1

```python
data['income'].unique()
data['income'].value_counts()

def salary_data(sal):
    if sal=='<=50K':
        return 0
    else:
        return 1
    
data['new_encoded_sal']=data['income'].apply(salary_data)
data.head(100)

```

```python
data['new_encoded_sal'].value_counts()
```
